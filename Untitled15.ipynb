```python
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
for column in numerical_columns:
    z_scores = np.abs(stats.zscore(df[column]))
    outliers = df[(z_scores > 3)]
    print(f"Outliers in column {column}:")
    print(outliers)
```
